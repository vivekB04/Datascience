```python
#data cleaning 
names=pd.Series(["Ashish","Rahul",None])
ages= pd.Series ([23,None,22])

#check missing
print(names.isnull())
print(ages.isnull())
print(names.notna())

#fill missing 
ages=ages.fillna(ages.mean())

#drop missing 
clean_df= pd.DataFrame({"name":names,"age": ages}).dropna()
print(clean_df)
```

```python
dept =pd.Series(["IT","HR","IT","HR"])
salary = pd.Series([30000,40000,60000,45000])

df=pd.DataFrame({"Dept":dept,"Salary":salary})
grouped = df.groupby("Dept")["Salary"].mean()
print(grouped)
```

```python
id1=pd.Series([1,2])
name=pd.Series(["Rahul","Bade"])

id2=pd.Series([1,2])
salary = pd.Series([30000,40000])

df1=pd.DataFrame({"ID":id1,"NAME":name})
df2=pd.DataFrame({"ID":id2,"Salary":salary})

merged=pd.merge(df1,df2,on="ID")
print(merged)
```
