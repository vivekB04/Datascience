```python
import matplotlib.pyplot as plt
```

```python
#Basic data
x=[1,2,3,4]
y=[10,20,30,40]

#simple line plot
plt.plot(x,y,marker='o')
plt.title("Basic Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#Line Plot
days = [1,2,3,4,5]
temperature =[30,32,35,33,31]

plt.plot(days,temperature,marker='o')

plt.title("Temperature over Days")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

```python
#Bar chart
names=["Ashish","Rahul","Kanha"]
marks=[85,90,78]

plt.bar(names,marks)
plt.show()
```

```python
# pie chart 
labels=["IT","HR","Sales"]
sizes=[40,30,30]

plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("Department Distribution")
plt.show()
```

```python
#Histogram
data=[10,20,20,30,30,30,40]
plt.hist(data,bins=4)
plt.title("Data Distribution")
plt.show()
```

```python
#Scatter plot
x=[1,2,3,4,5]
y=[5,7,9,11,13]
plt.scatter(x,y)
plt.title("Scatter plot")
plt.show()
```

```python
#Customizing PLot
x=[1,2,3]
y=[10,20,30]
plt.plot(x,y,linestyle='--',marker='o')

plt.title("Customized plot",fontsize=16)
plt.xlabel("X",fontsize=12)
plt.ylabel("Y",fontsize=12)

plt.grid(True)
plt.show()
```

```python
#Subplot
x = [1,2,3]
y = [1,4,9]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,[1,2,3])
plt.title("Plot 2")

plt.show()
```

```python
#Working with Figures and Axes
fig,ax=plt.subplots()
ax.plot([1,2,3],[10,20,30])
ax.set_title("00 Style plot")
plt.show()
```
